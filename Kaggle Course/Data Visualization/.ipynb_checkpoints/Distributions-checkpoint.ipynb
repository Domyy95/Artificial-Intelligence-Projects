```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Paths of the files to read
cancer_b_filepath = "./datasets/cancer_b.csv"
cancer_m_filepath = "./datasets/cancer_m.csv"

cancer_b_data = pd.read_csv(cancer_b_filepath, index_col="Id", parse_dates=True)
cancer_m_data = pd.read_csv(cancer_m_filepath, index_col="Id", parse_dates=True)

cancer_b_data.head()
```

```python
cancer_m_data.head()
```

```python
sns.distplot(a=cancer_b_data['Area (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label="Malignant", kde=False)
plt.legend()
```

```python
sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label="Benign")
sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label="Malignant")
```
